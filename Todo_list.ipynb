```python
def display_task():
    with open("tasks.txt","r") as f:
        l = f.readlines()
        print(l)
```

```python
def remove_task():
    new_list = []
    comp_task = input("Write the task you have completed: ")
    with open("tasks.txt","r") as f:
        tasks = f.readlines()
    for i in tasks:
        if i.strip()!=comp_task:
            new_list.append(i)
    with open("tasks.txt","w+") as f:
        f.writelines(new_list)
    print("Task has been removed")
    display_task()
```

```python
def add_task():
    no_of_tasks = int(input("Enter the no. of tasks you want to complete today:"))
    lst = []
    print("Add tasks to To-Do List:")
    for i in range(no_of_tasks):
        tasks = input()
        lst.append(tasks)
    with open("tasks.txt","a") as f:
        for i in lst:
            f.write(i+'\n')
    display_task()
```

```python
print("INSTRUCTIONS:\n 1.for adding a new task\n 2.for displaying your tasks list\n 3.for removing the completed task")
a="Y"
while a.upper() == 'Y':
    choice = int(input("Press a number"))
    if choice == 1:
        add_task()
    elif choice == 2:
        display_task()
    elif choice == 3:
        remove_task()
    else:
        print("Not Applicable")
    a = input("Do you want to still use it(Y/N)")
```
